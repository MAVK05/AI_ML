```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate some data using NumPy
# Create an array of 1000 evenly spaced points between 0 and 4*pi
x = np.linspace(0, 4 * np.pi, 1000)

# Create two slightly offset sine waves
y1 = np.sin(x)
y2 = np.sin(x + np.pi/4)

# 2. Plot the data using Matplotlib
plt.figure(figsize=(10, 5))

# Plot the lines with cool colors and labels
plt.plot(x, y1, color='cyan', linewidth=2, label='Sine Wave 1')
plt.plot(x, y2, color='magenta', linewidth=2, linestyle='--', label='Sine Wave 2 (offset)')

# Add a grid, title, and labels
plt.title("NumPy and Matplotlib Working Together! 🚀", fontsize=16)
plt.xlabel("X-axis (Time)")
plt.ylabel("Y-axis (Amplitude)")
plt.legend(loc="upper right")
plt.grid(True, linestyle=':', alpha=0.7)

# Set the background color to something dark like a premium UI
plt.gca().set_facecolor('#1e1e2e')
plt.gcf().patch.set_facecolor('#1e1e2e')
plt.tick_params(colors='white')
plt.title("NumPy and Matplotlib Working Together! 🚀", color='white', fontsize=16)
plt.xlabel("X-axis (Time)", color='white')
plt.ylabel("Y-axis (Amplitude)", color='white')
plt.legend(facecolor='#2d2d3f', edgecolor='none', labelcolor='white')

# Display the graph
plt.show()
```
